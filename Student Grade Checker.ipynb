```python
name = input("Enter your name : ")
marks_input = input("Enter your marks seperated by commas : ")
#split the string into a list
marks_list= marks_input.split(",")
#Convert each mark into integer and calculate total 
total = 0
for mark in marks_list:
    total +=int(mark)
#Calculate average    
    average= total/len(marks_list)
#Grading Logic    
if average >0:
    if(average<=40):
        print("Fail")
    elif(average<=60):
        print("Average")
    elif(average<=80):
        print("Pass")
    elif(average>=81):
        print("Excellent")
    else:
        print("Enter your valid inputs")
```
